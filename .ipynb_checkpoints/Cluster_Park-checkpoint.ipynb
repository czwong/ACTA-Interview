```python
# Dependencies
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Population and Area of Region in Square Feet
population = 1000
area = 10000

# Lets choose 5 coordinates that has shading along with the area they cover
shade_region = np.random.rand(5, 2)*area
shade_region_area = 1000
weather_hot = True
```

```python
shade_region
```

```python
# Generate Random Percentage of Population 
percentage_undertree = round(np.random.uniform(0.4,0.6),2)

# Find total population under shade
population_undertree = int(percentage_undertree * population)

if weather_hot:
    # Coordinates of population under shade
    X_1 = []
    for i in range(population_undertree):
        shade_point = random.choice(shade_region)
        X_1.append([shade_point[0] + random.uniform(-shade_region_area,shade_region_area), shade_point[1] + random.uniform(-shade_region_area,shade_region_area)])
    
    # Coordinates of remaining population
    X_2 = np.random.rand(population - population_undertree, 2) * area
    
    # Concatenate both arrays
    coords = np.concatenate((X_1, X_2), axis=0)
else:
    coords = np.random.rand(population, 2) * area
```

```python
# Plot and show scatter
plt.scatter(coords[:, 0], coords[:, 1])
plt.show()
```

```python
# Create a kmeans model using k = 6
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)

# Fit the model to the data
kmeans.fit(coords)

# Use the data to predict the clusters
predicted_clusters = kmeans.predict(coords)
```

```python
# Plot the clusters
plt.scatter(coords[:, 0], coords[:, 1], c=predicted_clusters, s=50, cmap='viridis')
plt.show()
```

```python
# Print the cluster centers and cluster labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
centers
```

```python
plt.scatter(coords[:, 0], coords[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

plt.show()
```

```python
sse = []
k_values = list(range(1, 20))

for k in k_values:
    km = KMeans(n_clusters=k)
    km.fit(coords)
    sse.append(km.inertia_)
    
# Plot sse against k
plt.figure(figsize=(6, 6))
plt.plot(k_values, sse, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance');
```

```python
kmeans = KMeans(n_clusters=5)

# Fit the model to the data
kmeans.fit(coords)

# Use the data to predict the clusters
predicted_clusters = kmeans.predict(coords)

centers = kmeans.cluster_centers_
labels = kmeans.labels_

plt.scatter(coords[:, 0], coords[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

plt.show()
```

```python
# Clustering Function
def cluster(pop, a, n, sa, c, d, weather_hot = True):
    # Population Count
    population = pop
    
    # Area of Region in Square Feet
    area = a

    # Lets choose n amount of coordinates that has shading along with the proximity that they cover
    shade_region = np.random.rand(n, 2)*area
    shade_region_area = sa
    
    # Generate random percentage of population looking for shade
    percentage_undertree = round(np.random.uniform(c,d),2)

    # Find total population under shade
    population_undertree = int(percentage_undertree * population)

    # Hot Weather Clustering
    if weather_hot:
        # Coordinates of population under shade
        X_1 = []
        for i in range(population_undertree):
            shade_point = random.choice(shade_region)
            X_1.append([shade_point[0] + random.uniform(-shade_region_area,shade_region_area), shade_point[1] + random.uniform(-shade_region_area,shade_region_area)])

        # Coordinates of remaining population
        X_2 = np.random.rand(population - population_undertree, 2) * area

        # Coordinate of each person
        coords = np.concatenate((X_1, X_2), axis=0)
        
    # Cold Weather Clustering
    else:
        coords = np.random.rand(population, 2) * area
        
    return coords
```

```python
coords = cluster(1000, 10000, 6, 1000, 0.4, 0.6, weather_hot = False)
```

```python
# Plot and show scatter
plt.scatter(coords[:, 0], coords[:, 1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=6)

# Fit the model to the data
kmeans.fit(coords)

# Use the data to predict the clusters
predicted_clusters = kmeans.predict(coords)

# Print the cluster centers and cluster labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

plt.scatter(coords[:, 0], coords[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

plt.show()
```
