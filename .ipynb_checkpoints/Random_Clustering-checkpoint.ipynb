```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Population and Area of Region in Square Feet
population = 1000
area = 10000
coords = np.random.rand(population, 2) * area
```

```python
# Plot and show scatter
plt.scatter(coords[:, 0], coords[:, 1])
plt.show()
```

```python
# Create a kmeans model using k = 12
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=30)

# Fit the model to the data
kmeans.fit(coords)

# Use the data to predict the clusters
# save the predictions as `predicted_clusters`
predicted_clusters = kmeans.predict(coords)
```

```python
# Plot the clusters
plt.scatter(coords[:, 0], coords[:, 1], c=predicted_clusters, s=50, cmap='viridis')
plt.show()
```

```python
# Print the cluster centers and cluster labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```
